```python
adult_ed=adult.drop('education',axis=1)
adult_ed.loc[(adult_ed.age=='#NAME?'),'age'] = np.nan
adult_ed.loc[(adult_ed.workclass=='?'),'workclass'] = np.nan
adult_ed.loc[(adult_ed.education_num=='#NAME?'),'education_num'] = np.nan
adult_ed.loc[(adult_ed.occupation=='?'),'occupation'] = np.nan
adult_ed.loc[(adult_ed.education_num=='?'),'education_num'] = np.nan
adult_ed.loc[(adult_ed.race=='#NAME?'),'race'] = np.nan

adult_ed.loc[(adult_ed.sex=='#NAME?'),'sex'] = np.nan
adult_ed.loc[(adult_ed.native_country=='?'),'native_country'] = np.nan
adult_ed.loc[(adult_ed.fnlwgt=='#NAME?'),'fnlwgt'] = np.nan
```

```python
adult_ed['age'] = adult_ed['age'].fillna(adult_ed['age'].mode()[0])
adult_ed['workclass'] = adult_ed['workclass'].fillna(adult_ed['workclass'].mode()[0])
adult_ed['education_num'] = adult_ed['education_num'].fillna(adult_ed['education_num'].mode()[0])
adult_ed['occupation'] = adult_ed['occupation'].fillna(adult_ed['occupation'].mode()[0])
adult_ed['education_num'] = adult_ed['education_num'].fillna(adult_ed['education_num'].mode()[0])
adult_ed['race'] = adult_ed['race'].fillna(adult_ed['race'].mode()[0])

adult_ed['sex'] = adult_ed['sex'].fillna(adult_ed['sex'].mode()[0])
adult_ed['native_country'] = adult_ed['native_country'].fillna(adult_ed['native_country'].mode()[0])
adult_ed['fnlwgt'] = adult_ed['fnlwgt'].fillna(adult_ed['fnlwgt'].mode()[0])
```
